# Enfierradura 	

La Nch 211 establece requisitos relacionados con el estado superficial de las barras, dimensiones, diámetros de doblado, tolerancias, toma de muestras para verificar el cumplimiento de los requisitos, empaque e identificación conforme a los planos estructurales de una obra de edificación ([EMB](http://www.emb.cl/construccion/articulo.mvc?xid=2949&ni=enfierradura-industrializada-en-obra-gruesa-altos-beneficios-constructivos) [Construcción](http://www.emb.cl/construccion/articulo.mvc?xid=2784&ni=norma-nch-211-requisitos-de-la-enfierradura-para-hormigon-armado), 2013).
