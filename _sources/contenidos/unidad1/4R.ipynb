# Reglamentaciones 

Solminiac & Thenoux (1998) menciona en su investigación que existen numerosas leyes, ordenanzas, reglas, reglamentos, normas y especificaciones que se deben seguir para llevar a cabo un proyecto de construcción.



<img src="../figuras/indice4R.png" width="1000">



## Leyes

Se establece por decreto supremo y son de obligado cumplimiento, los principales, referentes a la construcción, son las siguientes:

- Ley General de Urbanismo y construcción (DFL 458).

- Ley del Medio Ambiente (ley N°19300).

- Ley sobre contratos de trabajo y protección de los trabajadores (Ley N° 18372).

- Ley sobre accidentes y enfermedades profesionales (Ley N°16744).

- Otras (Código Civil, ley de la Propiedad Horizontal, ley de Copropiedad, Ley del IVA, DFL N°2).


## Ordenanzas


Cuerpos reglamentarios que apoyan a la ley general, son obligatorias.

- Ordenanza General de Urbanismo y Construcción (OGUC).


## Reglamentos


Ellos entre otras cosas, establecen requisitos administrativos, técnicos y de calidad que deben cumplirse en circunstancias particulares.

- Reglamento de instalaciones públicas.

- Reglamentos locales.

- Reglamento que rigen los contratos para la ejecución de obras en el sector de la vivienda y otras áreas.



## Normas


Conjunto de actividades que se documentan para establecer un orden, benefician a la comunidad y que esta haga el mejor uso de los recursos y satisfaga los requisitos funcionales y de seguridad.
Crean estándares de calidad y trabajan según modelos estandarizados.

Los establece el Instituto Nacional de Normalización (INN) de Chile, aunque ocasionalmente pueden provenir del exterior (AASHTO, DIN, ASTM, ACI, API, AISC, etc.).

Excepto cuando esté incluido en una ley u ordenanza, no se requiere su cumplimiento.
El INN cuenta con un listado de normas oficiales chilenas para la industria de la construcción (área F) que abarca el diseño general, arquitectónico, diseño, cálculo y ejecución de estructuras, acondicionamiento ambiental, seguridad, materiales y componentes, instalaciones, herramientas y equipos, y mobiliario. 



## Especificaciones técnicas


El Inspector de Obras (ITO) inspecciona la documentación relacionada con cada proyecto individual y que se utiliza para complementar los planos de ejecución.
