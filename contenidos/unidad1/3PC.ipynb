# Proyecto de Construcción


Peralta & Serpell (1991) Destaca en su investigación las etapas del proceso de trabajo para la realización de cualquier tipo de obra de construcción.

Solminiac & Thenoux (1998) Así mismo, menciona en su investigación las distintas etapas del proceso de obra, desde cómo se inicia hasta los documentos y permisos necesarios para su culminación.



<img src="../figuras/indicePC.png" width="1000">


## Definición de un proyecto de construcción

- Presencia de una necesidad. 

- Análisis.  

- Determinación de objetivos (sociales, económicos, funcionales, rentables). 

- Definición y evaluación de alternativas de solución (estudios de viabilidad, evaluación, financiación). 


## Diseño

- Analizar del terreno donde se construirá el proyecto (topografía, geología, hidrología, condiciones generales).

- Diseño arquitectónico.

- Diseño estructural.

- Análisis de impacto ambiental. 

- Diseño de las instalaciones.

- Redacción de documentos de licitación. 

- Constructibilidad y mantenimiento.


## Licitación

- Solicitar ofertas y adjudicar. 


## Construcción

- Definición de estrategia de gestión de la calidad.

- Obtención de permisos. 

- Redacción de bases administrativas incluyendo contratos, plazos, costos y relación entre propietarios y empresa contratistas.

- Planificación y programación de los trabajos (fijación de plazos parciales y globales, asignación de recursos). 

- Establecimiento de la metodología de trabajo.

- Estudio de presupuesto y costes de la obra. 

- Contrato de mano de obra. 

- Compra de materiales y/o alquiler de equipos.

- Materialización de la obra.

- Control. 

- Auditoria Ambiental.


## Puesta en Marcha

- Puesta en marcha.

- Controles mediante pruebas y ensayos para evaluar la calidad de la construcción. 

- análisis detallado de todos los materiales de construcción y acabados. 

- Pruebas de desempeño.

- Recepción ya sea preliminar, parcial o total.   

- Aprobación definitiva.


## Operación y mantenimiento

- Uso y mantenciones.


## Abandono

- Abandono y demolición.
